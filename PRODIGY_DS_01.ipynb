```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bank_data = pd.read_csv('bank.csv', sep=';')
print(bank_data.columns)
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=bank_data, x='housing', hue='loan')
plt.title('Comparison of Housing Loan and Personal Loan')
plt.xlabel('Housing Loan')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(bank_data['age'], bins=8, color='green', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age Distribution')
plt.show()
```

```python
education_value_count = bank_data['job'].value_counts()
education_value_count.plot(kind='bar')
plt.title('Count of Each Value in the Column')
plt.xlabel('Unique Values')
plt.ylabel('Count')
plt.show()
```

```python
marital_value_count = bank_data['marital'].value_counts()
marital_value_count.plot(kind='bar')
plt.title('Count of Each Value in the Column')
plt.xlabel('Unique Values')
plt.ylabel('Count')
plt.show()
```

```python
sns.barplot(data = bank_data, x='marital', y='age', hue='housing')
```

```python
sns.barplot(data = bank_data, x='education',y='age')
```

```python
housing_loan_yes = bank_data[bank_data['housing'] == 'yes']
housing_loan_no = bank_data[bank_data['housing'] == 'no']

age_distribution_yes = housing_loan_yes['age'].value_counts().sort_index()
age_distribution_no = housing_loan_no['age'].value_counts().sort_index()
```

```python
plt.plot(age_distribution_yes.index, age_distribution_yes.values, label='Housing Loan: Yes', color='blue')
plt.plot(age_distribution_no.index, age_distribution_no.values, label='Housing Loan: No', color='orange')

plt.title('Age Distribution for Housing Loan')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.show()
```
